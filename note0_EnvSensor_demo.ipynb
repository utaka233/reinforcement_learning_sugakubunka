# `EnvSensor` の環境デモ
module `environment.py` には、プラントのセンサーデータのシミュレータであるclass `EnvSensor` が準備されています。想定しているシチュエーションは
* 冷却装置のON/OFFを切り替えることで、センサーが観測している温度を調整する。

というものです。 class `EnvSensor` は、センサーが観測する温度の値に対応する `state` 変数を持ち、
* `reset()` : `state` に初期値を与えるmethod
* `step(action)` : `action` に応じて `state` を更新し、更新された `state` の値を返すmethod

を持っています。冷却装置ONが `action = 1`, OFFが `action = 0` に対応していると考えてください。

```python
# 必要なmoduleのimport
from environment import EnvSensor
import matplotlib.pyplot as plt
```

## A. `EnvSensor` の使い方

```python
# 環境の初期化
env = EnvSensor()
env.reset()
```

```python
# 行動後の状態
env.get_state(action = 0)
```

## B. シミュレーション
行動0を10回続けた場合、また行動1を10回続けた場合に環境がどのように変化するかを見てみましょう。

```python
# 行動0を10回続けた場合の状態変化（5 episodes）
n_episodes = 5
n_actions = 10
for i in range(n_episodes):
    env.reset()
    actions = [str(1+j) for j in range(n_actions)]
    states = [env.get_state(action = 0) for j in range(n_actions)]
    plt.plot(actions, states, label = str(i+1)+" episode")
plt.xlabel("number of actions")
plt.ylabel("state")
plt.legend(loc = "upper left")
plt.show()
```

```python
# 行動1を10回続けた場合の状態変化（5 episodes）
n_episodes = 5
n_actions = 10
for i in range(n_episodes):
    env.reset()
    actions = [str(1+j) for j in range(n_actions)]
    states = [env.get_state(action = 1) for j in range(n_actions)]
    plt.plot(actions, states, label = str(i+1)+" episode")
plt.xlabel("number of actions")
plt.ylabel("state")
plt.legend(loc = "upper right")
plt.show()
```
